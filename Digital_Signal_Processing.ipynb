```python
#CT Sinusoids Question-1

#construct and plot a CT sinosouid of frequency w = pi/6 rad/sec and phase phi = 0

step = 1000 #Resolution of time variable
t = np.linspace(0,60, step)

#parameters
omega = np.pi/6 #frequency
phi = 0 #phase difference
period = 2*(np.pi/omega) #time period

CTsin = np.sin(omega*t + phi) #ct signal

#plotting the sinosoid wrt time
plt.grid()#to show grid
plt.plot(t, CTsin)
plt.title('CT sinusoid')
plt.xlabel('TIME')
plt.ylabel('AMPLITUDE')
plt.show()#to plot individual graphs

#1.Is the constructed signal Periodic or Aperiodic. comment how and why?

print('Period of the sine wave is ', period)
print('\n')

#2.What is the relation between Phase () and Periodicity.

print('Phase and periodicity are independent of each other')

#non-zero phase plot
#plotting the sinosoid wrt time with non zero phase
phi = np.pi/2

sinosoidCT_phi = np.sin(omega*t + phi) #ct signal

plt.grid()
plt.plot(t, sinosoidCT_phi)
plt.title('CT sinosoid, phi = pi/2')
plt.xlabel('TIME')
plt.ylabel('AMPLITUDE')
plt.show()

print('Hence we can observe from the plots that the phase corresponded to no frequency change')
print('\n')
#3.What is the relation between Phase () and the Oscillations?

print('Plotting for a range of phase differences')
print('\n')
phi_sweep = np.array([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])

for phase_diff in phi_sweep:
    sinosoidCT_phase = np.sin(omega*t + phase_diff)
    plt.grid()
    plt.title('CT sinosoid')
    plt.xlabel('TIME')
    plt.ylabel('AMPLITUDE')
    plt.plot(t, sinosoidCT_phase)

plt.show()
print('Hence we can observe from the plots that the phase corresponded to no oscillations change')
print('\n')
#4.What is the significance of "step" in the following code above?

print('step refers to the resoultion of the time axis variable we are considering and will let us plot continous variable plots. Higher the resolution, more continuous the graph looks like.')
print('\n')
```

```python
#DFT
#4 point DFT for x[n]

def DFT(x): #function for DFT
    x = np.array(x)
    N = len(x)
    n = np.arange(0,N,1)
    nt = n.reshape((N, 1)) #taking transpose for dot product later
    e = np.exp(-2j * np.pi * nt * n / N) #each term summation
    Xk = np.dot(e, x) #second summation to get final array(DFT) of X[n]     
    return Xk

def IDFT(x): #function for IDFT
    x = np.array(x)
    N = len(x)
    n = np.arange(0,N,1)
    nt = n.reshape((N, 1)) #taking transpose for dot product later
    e = np.exp(2j * np.pi * k * n / N) #each term summation
    Xn = (1/N)*np.dot(e, x)#second summation to get final array(IDFT) of X[k]  
    return Xn

#input signal
xn = np.array([3, 1, 2, 1])
xk = DFT(xn)

print('4 point DFT of the signal is ', xk)

#comparision between the inbuilt function
m = sp.linalg.dft(4)
print('4 point DFT of the inbuilt function', m @ xn)

print('Thus both inbuilt function and created function gives almost the same results')
```
