```python
#PART B:
#1.a)
import numpy as np
import matplotlib.pyplot as plt
fs=int(input('Fs:'))
n=np.linspace(-10,10,20*fs)
m=[None]*(len(n))
for i in range(0,len(n)):
  if n[i]<0:
    m[i]=0
  else:
    m[i]=1
o=[None]*(len(n))
e=[None]*(len(n))
s=[None]*(len(n))
p=[None]*(len(n))
p[0]=(m[0]*m[0])
for i in range(0,len(n)):
  e[i]=(m[i]+m[len(n)-1-i])/2
  o[i]=((m[i]-m[len(n)-1-i])/2)
  s[i]=e[i]+o[i]
  if i>0:
    p[i]=p[i-1]+(m[i]*m[i])
ep=(p[len(n)-1])/(20*fs)  
print('energy:',p)
print('energy is increasing and is not finite')
print('power:',ep)
print('power is finite, thus, this is a power signal')   
plt.scatter(n,o,color='red')#odd component
plt.plot(n,o,color='red',linestyle='dashed')#dashed line also to shown to avoid overlap
plt.scatter(n,e,color='blue')#even component
plt.scatter(n,s,color='orange')#sum of components
plt.plot(n,s,color='orange',linestyle='dashed')#dashed line also to shown to avoid overlap
plt.scatter(n,m,color='green')#curve of the given function
print("As the orange and green curves overlap, we can conclude that the sum of even and odd component gives the original curve")
```

```python
#2.a)(i)
import numpy as np
import matplotlib.pyplot as plt
N=5
print('N:',N)
print('Input Signal x[k]=3k')
yf=[None]*21
yf2=[None]*21
n=np.arange(-10,11,1)
a=int(input('Scaling factor:'))
for z in range(0,len(n)):
  m=np.arange(z-N,z+N+1,1)
  x=[None]*(len(m))
  yf[z]=0
  for i in range(0,len(m)):
    x[i]=3*m[i]
    yf[z]=yf[z]+x[i]
  yf[z]=a*yf[z]
  yf2[z]=0
  for i in range(0,len(m)):
    x[i]=a*3*m[i]
    yf2[z]=yf2[z]+x[i]
plt.plot(n,yf2, linestyle='dashed')#New curve, dashed line to show overlap
plt.scatter(n,yf)#original curve*a
print('As the two curves are overlapping, we can say this system follows homogenous property')
```

```python
#2.a)(ii)
import numpy as np
import matplotlib.pyplot as plt
N=5
print('N:',N)
print('Input Signal-1 x[k]=a*k')
print('Input Signal-1 x[k]=b*k')
a=int(input('a:'))
b=int(input('b:'))
y1=[None]*21
y2=[None]*21
yf2=[None]*21
yf=[None]*21
n=np.arange(-10,11,1)
for z in range(0,len(n)):
  m=np.arange(z-N,z+N+1,1)
  x1=[None]*(len(m))
  x2=[None]*(len(m))
  xf=[None]*(len(m))
  y1[z]=0
  y2[z]=0
  yf2[z]=0
  yf[z]=0
  for i in range(0,len(m)):
    x1[i]=a*m[i]
    x2[i]=b*m[i]
    y1[z]=y1[z]+x1[i]
    y2[z]=y2[z]+x2[i]
    yf2[z]=y1[z]+y2[z]
  for i in range(0,len(m)):
    xf[i]=x1[i]+x2[i]
    yf[z]=yf[z]+xf[i]
plt.plot(n,yf2, linestyle='dashed')#New curve, dashed line to show overlap
plt.scatter(n,yf)#original curve
plt.scatter(n,y1)#y1 curve
plt.scatter(n,y2)#y2 curve
print('As the two curves are overlapping, we can say this system follows additive property')
print('Thus, we have shown Linearity Property')
```

```python
#2.b)
import numpy as np
import matplotlib.pyplot as plt
N=5
print('N:',N)
print('Input Signal x[k]=3k')
yf=[None]*21
yf2=[None]*21
yf3=[None]*21
n=np.arange(-10,11,1)
ts=int(input('Time shift:'))
for z in range(0,len(n)):
  m=np.arange(z-N,z+N+1,1)
  x=[None]*(len(m))
  yf[z]=0
  for i in range(0,len(m)):
    x[i]=3*m[i]
    yf[z]=yf[z]+x[i]
  yf2[z]=0
  for i in range(0,len(m)):
    x[i]=3*(m[i]-ts)  
    yf2[z]=yf2[z]+x[i]
  m2=np.arange(z-N-ts,z+N+1-ts,1)
  x2=[None]*(len(m2))
  yf3[z]=0
  for i in range(0,len(m2)):
    x2[i]=3*m2[i]
    yf3[z]=yf3[z]+x2[i]
plt.scatter(n,yf2)#shifted curve
plt.plot(n,yf)#original curve
plt.plot(n,yf3,linestyle='dashed')#new shifted curve, dashed line to show overlap
plt.plot([-9,-9+ts],[yf[1],yf3[1+ts]], linestyle='dashed')
print('As the two curves are overlapping, we can say this system follows Time-Invariance property')
```

```python
#2.d)
import numpy as np
import matplotlib.pyplot as plt
N=5
print('N:',N)
print('Input Signal: x[k]=3k*h(n-k) where h(a)=A*d(a). d(a) is the impulse discrete signal')
a=int(input('A:'))
n=int(input('n:'))
m=np.arange(n-N,n+N+1,1)
h=[None]*(len(m))
x=[None]*(len(m))
y=[None]*(len(m))
ysum=[None]*(len(m))
for i in range(0,len(m)):
  ysum[i]=0
  x[i]=3*m[i]
  for j in range(0,len(m)):
    if j==i:
      h[j]=a
    else:
      h[j]=0
  for z in range(0,len(m)):
    y[z]=x[i]*h[z]
    ysum[i]=ysum[i]+y[z]
print(x)
print(ysum)
plt.scatter(m,ysum)#output curve
plt.scatter(m,x)#input curve
plt.plot(m,ysum)#output curve
plt.plot(m,x)#input curve
print('As we can see in the graph, the output is proportional to just the present input, i.e y(n)=A*x(n)')
```

```python
#2.e)
import numpy as np
import matplotlib.pyplot as plt
N=5
print('N:',N)
print('Input Signal x[k]=3k')
print('h[k]=u[k]')
print('hi[k]=d(k)-d(k-1)')
print('h[k]*hi[k]=d(k), where d(k) is the impulse discrete signal')
n=int(input('n:'))
m=np.arange(n-N,n+N+1,1)
u=[None]*(len(m))
x=[None]*(len(m))
dk1=[None]*(len(m))
ysum=[None]*(len(m))
dk2=[None]*(len(m))
for i in range(0,len(m)):
  x[i]=m[i]*3
  for j in range(0,len(m)):
    if j==i:
      dk1[j]=1
    else:
      dk1[j]=0
    if j==i-1 and j>0:
      dk2[j]=1
    else:
      dk2[j]=0 
    if j<i:
      u[j]=0
    else:
      u[j]=1 
  ysum[i]=0
  for z in range(0,len(m)):
   ysum[i]=ysum[i]+(x[i]*u[z]*(dk1[z]-dk2[z]))
plt.scatter(m,ysum)#output curve
plt.plot(m,x,linestyle='dashed')#input curve, dashed line to show overlap
print('As we can see in the graph, both outputs are same')
print('Thus we have transformed the input signal back to its original signal')
```
