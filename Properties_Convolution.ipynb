```python
#1a
import numpy as np
import matplotlib.pyplot as plt
n=np.arange(-20,21,1)
d=[None]*len(n)
#delta function
for j in n:
  if j==4:
    d[j+20]=3
  else:
    d[j+20]=0
#unit step function(n-1)
u1=[None]*len(n)

for j in n:
  if j>=1:
    u1[j+20]=pow((3/4),j)
  else:
    u1[j+20]=0
#unit step function(n-5)
u2=[None]*len(n)

for j in n:
  if j>=5:
    u2[j+20]=pow((3/4),j)
  else:
    u2[j+20]=0


f1=[None]*len(n)
for j in n:
  f1[j+20]=u1[j+20]-u2[j+20]

#reversing the function
d2=[None]*len(n)
for j in n:
  d2[j+20]=d[40-j-20]


#convolution
y=[None]*len(n)
for j in n:
  sum=0
  temp_arr=[None]*len(n)
  if j>0:
    for k in range(0,j,1):
      temp_arr[k]=0
    for k in range(j,41,1):
      temp_arr[k]=d2[k-j]
    
    for k in range(0,41,1):
      sum+=temp_arr[k]*f1[k]
    y[j+20]=sum
  elif j<0:
    m=abs(j)
    for k in range(0,m,1):
      temp_arr[40-k]=0
    for k in range(0,41-m,1):
      temp_arr[k]=d2[k-j]
    
    for k in range(0,41,1):
      sum+=temp_arr[k]*f1[k]
    y[j+20]=sum
  else:
    for k in range(0,41,1):
      sum+=d2[k]*f1[k]
    y[j+20]=sum


plt.stem(n,y)
```
